```python
# Copia una Variable
nombre = 'Alexis'
nombre2 = nombre
nombre2 = 'Eduardo'

print(nombre, nombre2)
```

```python
numbers = ['one','two','three']
numbers_two = numbers

print(numbers, numbers_two)
numbers[0] = 'Alexis'
print(numbers, numbers_two, sep='\n')
```

```python
# Copiar listas
numbers = ['one','two','three']
numbers_copy = numbers.copy()

numbers[0] = 'Alexis'
print(numbers, numbers_copy, sep='\n')
```

```python
# Otra forma de copiar listas
numbers = ['one','two','three']
copy = numbers[:]

numbers[0] = 'Alexis'
print(numbers, copy, sep='\n')
```

```python
# Ciclo
numbers_one = [1,2,3]
numbers_two = [4,5,6]

for i in numbers_two:
  numbers_one.append(i)

print(numbers_one)
```

```python
# Merge List
numbers_one = [1,2,3]
numbers_two = [4,5,6]

numbers = numbers_one + numbers_two
numbers_one.extend(numbers_two)

print(numbers)
print(numbers_one)
```

```python
numbers_one = [1,2,3]
numbers_two = [4,5,6]

for i in (numbers_one, numbers_two):
  print(i)

numbers = []
for i in (numbers_one, numbers_two):
  for j in i:
    numbers.append(j)

print(numbers)
```

```python
# List Comprehesion

numbers = [j for i in (numbers_one, numbers_two) for j in i]
print(numbers)
```

```python
numbers_one = [1,2,3]
numbers_two = [4,5,6]

numbers = [*numbers_one, *numbers_two]
print(numbers)
```
