**Slicing List**

```python
players=['one','two','thre','four','five']
print(players)
print(type(players))
print (players[1:3])
```

```python
print(players[0:1])
print(players[:3])
print(players)
```

```python
print(players[2:])
```

```python
print(players[:])
```

```python
print(players[-1:2])
```

```python
print(players[-1])
print(players[:-2])
print(players[-3:])
print (players[-5:-2])
```

```python
first_index=-5
print(players[first_index:])
```

```python
##palindormo = oso
players.append('six')
players.append('seven')
players.append('eight')
players.append('nine')
players.append('ten')
```

```python
print(players)
print(players[2:9:2])
print(players[::2])
```

```python
number= list(range(1,11))
print(number)
print(len(number))
print(number[::-1])
```
